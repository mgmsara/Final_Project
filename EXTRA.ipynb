```python
# sleep function 
from random import randint
def sleep_loop_playlist():
    from time import sleep
    for i in range(5):
        print(i)
        wait_time = randint(1,4)
        print("I will sleep for " + str(wait_time) + " seconds.")
        sleep(wait_time)
```
