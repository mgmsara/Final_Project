```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime
from iexfinance.stocks import Stock
from iexfinance.stocks import get_historical_data
```

### Create a dataframe from IEX API 

```python
#DataFrame of Tesla stock price in 2022
start = datetime(2022, 1, 1)
end = datetime(2022, 12, 1)
tesla_df = get_historical_data("TSLA", start, end, output_format='pandas',token="sk_e0179f146033476aaf5291091322c825")
```

```python
tesla_df
```

```python
#Keeping only useful columns
tesla_df= tesla_df[['close','high','low','open','priceDate']]
tesla_df = tesla_df.rename({'priceDate':'date'}, axis=1)
tesla_df
```

```python
#plot of Tesla close stock value

plt.figure(figsize=(10,10))
plt.plot(tesla_df.index, tesla_df['close'])
plt.xlabel("date")
plt.ylabel("$ price")
plt.title("Tesla Stock Price 1.01.22 - 1.12.22");
```

```python
#Save file to csv
tesla_df.to_csv('Data/tesla_stock2022.csv')
```
