```python
import pandas as pd
import re
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv('FINAL.csv')
```

```python
# Cleaning Elon Musk Perception
```

```python
df
```

```python
list(df['location'])
```

```python
df['Country'] = df['location']

# define the function that will clean the location column
def clean_location(x):
    if isinstance(x, str):
        # remove emoji using regular expressions
        x = re.sub(r'[^\w\s]', '', x)
        # remove leading and trailing white space
        x = x.strip()
        return x
    return None

# apply the function to the location column
df["Country"] = df["Country"].apply(clean_location)

# drop rows with NaN values in the location column
#df.dropna(subset=["Country"], inplace=True)
df['Country']
```

```python
# create the Nominatim object
geolocator = Nominatim(user_agent="my_app")

# define the function that will get the country from the location
def get_country(x):
    if isinstance(x, str) and x != 'None':
        # use the Nominatim object to get the location's coordinates
        try:
            location = geolocator.geocode(x)
            if location:
                # return the country name from the location object
                country = location.raw["display_name"].split(", ")[-1]
                return country
        except Exception:
            return None
    return None

# apply the function to the location column and store the result in the new 'Country' column
df["Country_real"] = df["Country"].apply(get_country)
df["Country_real"]
    
```

```python
df["Country_real"].unique()
```

```python
df["Country_real"].value_counts()
```

```python
def extract_compound(x):
    return eval(x)['compound']

df['Compound'] = df["score"].apply(extract_compound)
```

```python
df
```

```python
df.to_csv('FINAL_NEW.csv')
```
